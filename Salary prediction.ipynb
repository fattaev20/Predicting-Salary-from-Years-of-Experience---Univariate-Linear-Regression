```python
# Import required libraries
import pandas as pd                  # For reading and processing CSV data
import matplotlib.pyplot as plt      # For plotting graphs
```

```python
# Load dataset from CSV file
df = pd.read_csv('Salary_dataset.csv')

# Extract relevant columns into separate variables
years = df['YearsExperience']   # Independent variable (X)
salary = df['Salary']           # Dependent variable (Y)
```

```python
# Plot initial scatterplot of Salary vs. Years of Experience
# plt.figure(figsize=(12, 8))   # Optional: adjust figure size
plt.scatter(salary, years, marker="x", color='r')  # Red 'x' markers
plt.xlabel("Salary")
plt.ylabel("Years of Experience")
# plt.xticks(range(30000, 130000, 5000))  # Optional custom x-axis ticks
# plt.yticks(range(0, 12, 1))             # Optional custom y-axis ticks
```

```python
# Prepare variables for training
x_train = years
y_train = salary

# Initial parameters for linear regression
w_init = 0    # Initial weight (slope)
b_init = 0    # Initial bias (intercept)

# Number of training examples
size = len(x_train)
```

```python
# Cost function for Linear Regression (Mean Squared Error)
def calculate_cost(x, y, w, b):
    cost = 0
    for i in range(size):
        cost = cost + (x[i] * w + b - y[i]) ** 2   # Sum of squared errors
    return cost

# Calculate initial cost with w=0, b=0
j = calculate_cost(x_train, y_train, w_init, b_init) / (2 * size)
```

```python
# Learning rate for Gradient Descent
alpha = 0.01

# Function to compute total derivative with respect to w
def calculate_total_dw(x, y, w, b):
    total_dw = 0
    for i in range(size):
        total_dw = total_dw + (w * x[i] + b - y[i]) * x[i]
    return total_dw

# Function to compute total derivative with respect to b
def calculate_total_db(x, y, w, b):
    total_db = 0
    for i in range(size):
        total_db = total_db + (w * x[i] + b - y[i])
    return total_db
```

```python
# Gradient Descent algorithm to find optimal w and b
for i in range(15001):
    # Update bias
    b = b_init - alpha * (calculate_total_db(x_train, y_train, w_init, b_init) / size)
    # Update weight
    w = w_init - alpha * (calculate_total_dw(x_train, y_train, w_init, b_init) / size)

    # Store updated parameters for next iteration
    w_init = w
    b_init = b

    # Print progress every 1000 iterations
    if i % 1000 == 0:
        print(f"{i}th iteration: w:{w} / b:{b}")
```

```python
# Generate predictions using the trained model
predicted_y = []
for i in range(size):
    predicted_y.append(w_init * x_train[i] + b_init)
```

```python
# Plot actual data and regression line
plt.scatter(salary, years, marker="x", color='r')  # Actual data
plt.plot(predicted_y, years)                       # Regression line
plt.xlabel("Salary")
plt.ylabel("Years of Experience")
```
